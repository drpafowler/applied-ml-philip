```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="ticks")
plt.style.use('seaborn-v0_8-ticks')
```

```python
# Load the CSV file into a DataFrame
datasaurus_df = pd.read_csv('datasaurus.csv')

# Display the first few rows of the DataFrame
datasaurus_df.head()
```

```python
unique_datasets = datasaurus_df['dataset'].unique()
print(unique_datasets)
```

```python
# Filter the DataFrame for the 'dino' dataset
dino_data = datasaurus_df[datasaurus_df['dataset'] == 'dino']

# display statistics
print(dino_data.describe())

# Create a scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dino_data, x='x', y='y')
plt.title('Scatterplot for Dataset: dino')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Filter the DataFrame for the 'x_shape' dataset
x_shape_data = datasaurus_df[datasaurus_df['dataset'] == 'x_shape']

# display statistics
print(x_shape_data.describe())

# Create a scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=x_shape_data, x='x', y='y')
plt.title('Scatterplot for Dataset: x_shape')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
