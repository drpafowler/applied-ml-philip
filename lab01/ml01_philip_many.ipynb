### 4.4 Train the polynomial regression model

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Create a polynomial regression model
degree = 2
poly = make_pipeline(PolynomialFeatures(degree), LinearRegression())
poly.fit(X_train, y_train)  
```

### 4.5 Report R^2, MAE, RMSE for the polynomial model

```python
# Calculate R^2 for polynomial regression
y_pred_poly = poly.predict(X_test)
r2_poly = r2_score(y_test, y_pred_poly)
print(f'R²: {r2_poly:.2f}')
```

```python
# Calculate the mean absolute error for polynomial regression
mae_poly = mean_absolute_error(y_test, y_pred_poly)
print(f'Mean Absolute Error: {mae_poly:.2f}')
```

```python
# Calculate the root mean squared error for polynomial regression
rmse_poly = root_mean_squared_error(y_test, y_pred_poly)
print(f'Root Mean Squared Error: {rmse_poly:.2f}')
```

### 4.6 Train the Random Forest regression model

```python
# create a random forest regression model
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor(n_estimators=100, random_state=dont_panic)
random_forest.fit(X_train, y_train)
```

### 4.7 Report R^2, MAE, RMSE for the Random Forest model

```python
# calculate R^2 for random forest regression
y_pred_rf = random_forest.predict(X_test)
r2_rf = r2_score(y_test, y_pred_rf)
print(f'R²: {r2_rf:.2f}')
```

```python
# calculate the mean absolute error for random forest regression
mae_rf = mean_absolute_error(y_test, y_pred_rf)
print(f'Mean Absolute Error: {mae_rf:.2f}')
```

```python
# calculate the root mean squared error for random forest regression
rmse_rf = root_mean_squared_error(y_test, y_pred_rf)
print(f'Root Mean Squared Error: {rmse_rf:.2f}')
```

### 4.8 Train the Gradient Boosting regression model

```python
# create a gradient boosting regression model
from sklearn.ensemble import GradientBoostingRegressor
gradient_boosting = GradientBoostingRegressor(n_estimators=100, random_state=dont_panic)
gradient_boosting.fit(X_train, y_train)
```

### 4.9 Report R^2, MAE, RMSE for the Gradient Boosting model

```python
# calculate R^2 for gradient boosting regression
y_pred_gb = gradient_boosting.predict(X_test)
r2_gb = r2_score(y_test, y_pred_gb)
print(f'R²: {r2_gb:.2f}')
```

```python
# calculate the mean absolute error for gradient boosting regression
mae_gb = mean_absolute_error(y_test, y_pred_gb)
print(f'Mean Absolute Error: {mae_gb:.2f}')
```

```python
# calculate the root mean squared error for gradient boosting regression
rmse_gb = root_mean_squared_error(y_test, y_pred_gb)
print(f'Root Mean Squared Error: {rmse_gb:.2f}')
```

### 4.10 Train the KNN regression model

```python
# create a knn regression model
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)
```

### 4.11 Report R^2, MAE, RMSE for the KNN model

```python
# calculate R^2 for knn regression
y_pred_knn = knn.predict(X_test)
r2_knn = r2_score(y_test, y_pred_knn)
print(f'R²: {r2_knn:.2f}')
```

```python
# calculate the mean absolute error for knn regression
mae_knn = mean_absolute_error(y_test, y_pred_knn)
print(f'Mean Absolute Error: {mae_knn:.2f}')
```

```python
# calculate the root mean squared error for knn regression
rmse_knn = root_mean_squared_error(y_test, y_pred_knn)
print(f'Root Mean Squared Error: {rmse_knn:.2f}')
```

## Section 5. Model Comparison and Selection
### 5.1 Compare the models

```python
# Create a dictionary with model names and their corresponding R² values
r2_scores = {
    'Linear Regression': r2_linear,
    'Polynomial Regression': r2_poly,
    'Random Forest': r2_rf,
    'Gradient Boosting': r2_gb,
    'KNN': r2_knn
}

# Create a bar plot to compare the R² values
plt.figure(figsize=(10, 6))
plt.bar(r2_scores.keys(), r2_scores.values())
plt.xlabel('Model')
plt.ylabel('R² Score')
plt.title('Comparison of R² Scores for Different Models(Bigger is better)')
plt.ylim(0, 1)
plt.show()
```

```python
# Create a dictionary with model names and their corresponding MAE values
mae_scores = {
    'Linear Regression': mae_linear,
    'Polynomial Regression': mae_poly,
    'Random Forest': mae_rf,
    'Gradient Boosting': mae_gb,
    'KNN': mae_knn
}

# Create a bar plot to compare the MAE values
plt.figure(figsize=(10, 6))
plt.bar(mae_scores.keys(), mae_scores.values())
plt.xlabel('Model')
plt.ylabel('Mean Absolute Error')
plt.title('Comparison of MAE for Different Models (Smaller is Better)')
plt.show()
```

```python
# Create a dictionary with model names and their corresponding RMSE values
rmse_scores = {
    'Linear Regression': rmse_linear,
    'Polynomial Regression': rmse_poly,
    'Random Forest': rmse_rf,
    'Gradient Boosting': rmse_gb,
    'KNN': rmse_knn
}

# Create a bar plot to compare the RMSE values
plt.figure(figsize=(10, 6))
plt.bar(rmse_scores.keys(), rmse_scores.values())
plt.xlabel('Model')
plt.ylabel('Root Mean Squared Error')
plt.title('Comparison of RMSE for Different Models (Smaller is Better)')
plt.show()
```
